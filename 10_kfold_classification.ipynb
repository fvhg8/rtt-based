```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import pickle
```

```python
dataset = pd.read_csv('Dataset_TA.csv')
```

```python
rtt = dataset[['rtt1','rtt2','rtt3','rtt4','rtt5','rtt6']].to_numpy()
gt_coordinates_x = dataset['gtx'].to_numpy()
gt_coordinates_y = dataset['gty'].to_numpy()
pos_id = dataset['PosId'].to_numpy()
ap1x = dataset['ap1x'].to_numpy()
ap1y = dataset['ap1y'].to_numpy()
ap2x = dataset['ap2x'].to_numpy()
ap2y = dataset['ap2y'].to_numpy()
ap3x = dataset['ap3x'].to_numpy()
ap3y = dataset['ap3y'].to_numpy()
ap4x = dataset['ap4x'].to_numpy()
ap4y = dataset['ap4y'].to_numpy()
ap5x = dataset['ap5x'].to_numpy()
ap5y = dataset['ap5x'].to_numpy()
ap6x = dataset['ap6x'].to_numpy()
ap6y = dataset['ap6x'].to_numpy()
```

```python
X = rtt
y = pos_id
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=900)
kfold_svm = StratifiedKFold(n_splits=3, shuffle=True, random_state=750)
```

# Knn

```python
from sklearn.neighbors import KNeighborsClassifier
results_knn = cross_val_score(KNeighborsClassifier(), X, y,cv=kfold)
```

```python
print(results_knn.mean())
print(results_knn)
```

```python
with open('10_fold_knn.pkl','wb') as pickle_file:
    pickle.dump(results_knn,pickle_file)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators = 10, criterion = 'gini', random_state = 500)
results_forest = cross_val_score(random_forest, X, y,cv=kfold)
```

```python
print(results_forest.mean())
print(results_forest)
```

```python
with open('10_fold_random_forest.pkl','wb') as pickle_file:
    pickle.dump(results_forest,pickle_file)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(criterion = 'gini', random_state = 385)
results_tree = cross_val_score(decision_tree, X, y,cv=kfold)
```

```python
print(results_tree.mean())
print(results_tree)
```

```python
with open('10_fold_decision_tree.pkl','wb') as pickle_file:
    pickle.dump(results_tree,pickle_file)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
naive_bayes= GaussianNB()
results_bayes = cross_val_score(naive_bayes, X, y,cv=kfold)
```

```python
print(results_bayes.mean())
print(results_bayes)
```

```python
with open('10_fold_naive_bayes.pkl','wb') as pickle_file:
    pickle.dump(results_bayes,pickle_file)
```

# SVM

```python
from sklearn.svm import SVC
svm = SVC(kernel = 'linear', random_state = 90)
results_svm = cross_val_score(svm, X, y,cv=kfold_svm)
```

```python
print(results_svm.mean())
print(results_svm)
```
